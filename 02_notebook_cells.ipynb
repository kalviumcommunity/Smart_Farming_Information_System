# Understanding Notebook Cells: Code vs Markdown

This notebook demonstrates the difference between Code and Markdown cells in a Jupyter Notebook, and how to use them together to create a readable and execution-ready document.

## 1. Using Code Cells for Execution

The cell below is a **Code cell**. It contains Python code that can be executed by the kernel. Here, we will define a simple variable and print it.

```python
greeting = "Hello, Jupyter Notebook!"
print(greeting)
```

## 2. Using Markdown Cells for Explanation

As seen above, this cell is a **Markdown cell**. It is used to format text, add headings, bullet points, and explain the logic of the code without executing anything.

In the next code cell, we will perform a basic mathematical operation to show how code cells handle computation.

```python
a = 15
b = 27
result = a + b
print(f"The sum of {a} and {b} is {result}")
```
