## 🔧 Step 1: Import Libraries

In this step, we import the required libraries for data handling and basic analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 📊 Step 2: Create Sample Data

To simulate real-world data, we create a simple dataset representing crop prices.

```python
data = {
    "Crop": ["Wheat", "Rice", "Maize"],
    "Price": [2000, 1800, 1500]
}

df = pd.DataFrame(data)
df
```

## 📌 Understanding the Data

The dataset contains:

- Crop names
- Their corresponding market prices

This helps us analyze which crop is more profitable.

## 📈 Step 3: Basic Analysis

We calculate:

1. Average price
2. Maximum price
3. Minimum price

```python
avg_price = df["Price"].mean()
max_price = df["Price"].max()
min_price = df["Price"].min()

avg_price, max_price, min_price
```

## 📊 Results Interpretation

From the analysis:

- The average crop price gives an overall market idea
- The maximum price shows the most profitable crop
- The minimum price indicates lower-value crops

These insights help farmers choose better crops.
